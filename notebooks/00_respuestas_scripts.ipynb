## PARTE 2
### ¿Cómo cambia el error al aumentar el ángulo?

El error aumenta significativamente. Al incrementar el ángulo inicial (de 5° a 60°), la aproximación lineal ($\sin(\theta) \approx \theta$) no cumple, haciendo que la diferencia entre el modelo real (no lineal) y el simplificado sea cada vez mayor.

### ¿El error crece en el tiempo?

Sí. Debido a que ambos modelos tienen períodos ligeramente distintos, las señales se van desfasando conforme avanza el tiempo. Esto causa que la diferencia acumulada (el error) oscile con una amplitud cada vez más grande.

### ¿Qué ocurre con el período?
- Modelo Lineal: El período se mantiene constante (no depende del ángulo).

- Modelo No Lineal: El período aumenta a medida que el ángulo inicial es mayor. El péndulo "real" tarda más tiempo en completar una oscilación cuando la amplitud es grande.

## PARTE 3
### Explica el desfase entre modelos

El desfase ocurre porque ambos modelos calculan períodos ligeramente diferentes. Al no oscilar exactamente a la misma velocidad, las ondas se van desalineando con el paso del tiempo.

### ¿Qué significa que la frecuencia dependa de la amplitud?

Significa que el sistema es no lineal. En un sistema lineal (como el péndulo ideal de ángulos pequeños), la frecuencia es constante sin importar la fuerza del impulso. Aquí, el ángulo inicial altera físicamente el tiempo que tarda en completar un ciclo: a mayor amplitud, menor frecuencia.
